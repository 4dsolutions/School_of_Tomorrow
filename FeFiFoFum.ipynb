```python
import sympy as sy
from tetravolume import E, Evol, root2, root5
from flextegrity import PD
```

```python
evol = E().ivm_volume().simplify() # using Piero type edge inputs
evol
```

```python
Evol # directly in terms of phi
```

```python
sy.Eq(Evol, evol) # same result
```

```python
φ = sy.Symbol('φ') # shorthand
```

```python
φ = (1 + root5)/2
φ
```

```python
# Dissecting the E-mod
Fe  = (root2/8) * φ**-7
Fi  = (root2/8) * φ**-6
Fo  = (root2/8) * ((5 - root5)/5) * (φ**-4)
Fum = (root2/8) * (root5/5) * (φ**-4)
```

```python
(Fe + Fi).simplify()
```

```python
sy.Eq(Fe + Fi, (root2/8) * φ**-5) # confirm 
```

```python
sy.Eq(Fo + Fum, (root2/8) * φ**-4) # confirm
```

```python
sy.Eq(Fe + Fi + Fo + Fum, Evol) # all add to E volume
```

In mentioning the PD (pentagonal dodecahedron), DK means the PD that goes along with the E mods RT, not the SuperRT of long diagonals D.

The SuperRT scaled down by 1/φ per all edges is the Emods RT, to which there correspond a PD and Icosa shrunk by the same scale factor. 

That's the PD we're talking about: the default PD (paired with D-edged icosa) scaled down by φ to the 3rd (volume-wise).

```python
pd = PD() * (1/φ)
pd.volume
```

```python
120 * (Evol - Fum)
```

```python
pd.volume.evalf()
```

```python
(120 * (Evol - Fum)).evalf()
```

```python
sy.Eq(120 * (Evol - Fum), pd.volume)
```

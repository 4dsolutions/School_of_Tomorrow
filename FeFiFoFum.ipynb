```python
import sympy as sy
from tetravolume import E, Evol, root2, root5
```

```python
evol = E().ivm_volume().simplify() # using Piero type edge inputs
evol
```

```python
Evol # directly in terms of phi
```

```python
sy.Eq(Evol, evol) # same result
```

```python
φ = sy.Symbol('φ') # shorthand
```

```python
φ = (1 + root5)/2
φ
```

```python
# Dissecting the E-mod
Fe  = (root2/8) * φ**-7
Fi  = (root2/8) * φ**-6
Fo  = (root2/8) * ((5 - root5)/5) * (φ**-4)
Fum = (root2/8) * (root5/5) * (φ**-4)
```

```python
(Fe + Fi).simplify()
```

```python
sy.Eq(Fe + Fi, (root2/8) * φ**-5) # confirm 
```

```python
sy.Eq(Fo + Fum, (root2/8) * φ**-4) # confirm
```

```python
sy.Eq(Fe + Fi + Fo + Fum, Evol) # all add to E volume
```

This last slide shows off the tiny difference by allowing the a red unit-radius ball to protrude through the RT skin slightly, when the RT shrinks to 120 T size.  How do we get there using [our Python Framework](Platonics.ipynb)?

The framework begets an RT that is initially, by default, what we sometimes call the SuperRT: that which is begotten by the default Icosahedron and its dual, edges crossing, the default Pentagonal Dodecahedron. This Icosahedron has edges = D, the diameter of an IVM ball, our reference length. 

To shrink it down to the RT made of Es, we first need to scale all its edges by $1/\phi$. Volume will thereby shrink by $1/\phi^{3}$.

The Icosahedron may be comprised of three mutually orthogonal golden rectangles, showing how center-to-mid-edge would be radius $\phi$ assuming edges = 1. Icosahedron mid-edges are likewise at RT face centers, where they cross the edges of the Icosa's dual.

Then, to shrink it down even more, to the RT made of Ts, we multiply by 0.99948333226... or, algebraically: $(2/3)^{1/3}(\phi/\sqrt{2})$. Lets combine those two steps and enlist the support of Google Gemini to distill the result into a simplest expression. 

## Prompt to Gemini:  

Simplify (1/phi)*(2/3)^(1/3)*(phi/sqrt(2)) where phi = (1 + sqrt(5))/2

## Gemini's reply:

The simplified form of the expression is $\frac{1}{\sqrt[6]{18}}$ or $\frac{\sqrt[6]{432}}{6}$. [1] 
## Step-by-Step Simplification

   1. Cancel the Golden Ratio ($\phi$) Terms
   Observe that the expression contains both $\frac{1}{\phi}$ and $\phi$ as factors. These two terms multiply to $1$, allowing them to be removed from the expression entirely:
   $$\left(\frac{1}{\phi}\right) \cdot \left(\frac{2}{3}\right)^{1/3} \cdot \left(\frac{\phi}{\sqrt{2}}\right) = \frac{\phi}{\phi} \cdot \left(\frac{2}{3}\right)^{1/3} \cdot \frac{1}{\sqrt{2}} = \left(\frac{2}{3}\right)^{1/3} \cdot \frac{1}{\sqrt{2}}$$ 
   2. Express All Terms with Rational Exponents
   To combine the roots, rewrite $\sqrt{2}$ as $2^{1/2}$ and the cube root as a $1/3$ power:
   $$\frac{2^{1/3}}{3^{1/3} \cdot 2^{1/2}}$$ 
   3. Combine the Base 2 Terms
   Apply the law of exponents ($\frac{x^a}{x^b} = x^{a-b}$) to the terms with base 2:
   $$\frac{2^{1/3 - 1/2}}{3^{1/3}} = \frac{2^{-1/6}}{3^{1/3}}$$ 
   4. Convert to a Single Radical
   To bring both terms under a single $1/6$ root (the sixth root), rewrite $3^{1/3}$ as $3^{2/6}$:
   $$\frac{1}{2^{1/6} \cdot 3^{2/6}} = \frac{1}{(2 \cdot 3^2)^{1/6}} = \frac{1}{(2 \cdot 9)^{1/6}} = \frac{1}{18^{1/6}}$$ 
   5. Rationalize the Denominator (Optional)
   Multiply the numerator and denominator by $18^{5/6}$ to remove the radical from the bottom:
   $$\frac{18^{5/6}}{18} = \frac{\sqrt[6]{18^5}}{18} = \frac{\sqrt[6]{1889568}}{18} = \frac{\sqrt[6]{432}}{6}$$ [1, 2, 3] 

## ✅ Final Answer
The simplified expression is $\frac{1}{\sqrt[6]{18}}$, which is approximately $0.6177$. [1]

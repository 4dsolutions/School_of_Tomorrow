This last slide shows off the tiny difference by allowing the a red unit-radius ball to protrude through the RT skin slightly, when the RT shrinks to 120 T size.  How do we get there using [our Python Framework](Platonics.ipynb)?

The framework begets an RT that is initially, by default, what we sometimes call the SuperRT: that which is begotten by the default Icosahedron and its dual, edges crossing, the default Pentagonal Dodecahedron. This Icosahedron has edges = D, the diameter of an IVM ball, our reference length. 

To shrink it down to the RT made of Es, we first need to scale all its edges by $1/\phi$. Volume will thereby shrink by $1/\phi^{3}$.

The Icosahedron may be comprised of three mutually orthogonal golden rectangles, showing how center-to-mid-edge would be radius $\phi$ assuming edges = 1. Icosahedron mid-edges are likewise at RT face centers, where they cross the edges of the Icosa's dual.

Then, to shrink it down even more, to the RT made of Ts, we multiply by 0.99948333226... or, algebraically: $(2/3)^{1/3}(\phi/\sqrt{2})$. Lets combine those two steps and enlist the support of Google Gemini to distill the result into a simplest expression. 

## Prompt to Gemini:  

Simplify (1/phi)*(2/3)^(1/3)*(phi/sqrt(2)) where phi = (1 + sqrt(5))/2

## Gemini's reply:

The simplified form of the expression is $\frac{1}{\sqrt[6]{18}}$ or $\frac{\sqrt[6]{432}}{6}$. [1] 
## Step-by-Step Simplification

   1. Cancel the Golden Ratio ($\phi$) Terms
   Observe that the expression contains both $\frac{1}{\phi}$ and $\phi$ as factors. These two terms multiply to $1$, allowing them to be removed from the expression entirely:
   $$\left(\frac{1}{\phi}\right) \cdot \left(\frac{2}{3}\right)^{1/3} \cdot \left(\frac{\phi}{\sqrt{2}}\right) = \frac{\phi}{\phi} \cdot \left(\frac{2}{3}\right)^{1/3} \cdot \frac{1}{\sqrt{2}} = \left(\frac{2}{3}\right)^{1/3} \cdot \frac{1}{\sqrt{2}}$$ 
   2. Express All Terms with Rational Exponents
   To combine the roots, rewrite $\sqrt{2}$ as $2^{1/2}$ and the cube root as a $1/3$ power:
   $$\frac{2^{1/3}}{3^{1/3} \cdot 2^{1/2}}$$ 
   3. Combine the Base 2 Terms
   Apply the law of exponents ($\frac{x^a}{x^b} = x^{a-b}$) to the terms with base 2:
   $$\frac{2^{1/3 - 1/2}}{3^{1/3}} = \frac{2^{-1/6}}{3^{1/3}}$$ 
   4. Convert to a Single Radical
   To bring both terms under a single $1/6$ root (the sixth root), rewrite $3^{1/3}$ as $3^{2/6}$:
   $$\frac{1}{2^{1/6} \cdot 3^{2/6}} = \frac{1}{(2 \cdot 3^2)^{1/6}} = \frac{1}{(2 \cdot 9)^{1/6}} = \frac{1}{18^{1/6}}$$ 
   5. Rationalize the Denominator (Optional)
   Multiply the numerator and denominator by $18^{5/6}$ to remove the radical from the bottom:
   $$\frac{18^{5/6}}{18} = \frac{\sqrt[6]{18^5}}{18} = \frac{\sqrt[6]{1889568}}{18} = \frac{\sqrt[6]{432}}{6}$$ [1, 2, 3] 

## ✅ Final Answer
The simplified expression is $\frac{1}{\sqrt[6]{18}}$, which is approximately $0.6177$. [1]

## A Fourth RT

Another RT we have so far not mentioned is the one with volume 7.5, that shares vertices with the RD (rhombic dodecahedron) of volume 6. An easy way to get there is from the RT_T "station stop". 7.5 is simple (3/2) * 5, and the RT_T is volume 5. Since we use a linear scalefactor to achieve 3rd powering effects, we'll need a scale factor that's the 3rd root of (3/2).

<a data-flickr-embed="true" href="https://www.flickr.com/photos/kirbyurner/55213393556/in/dateposted-public/" title="From Cascadian Synergetics"><img src="https://live.staticflickr.com/65535/55213393556_12abc0383b_z.jpg" width="640" height="362" alt="From Cascadian Synergetics"/></a><script async src="//embedr.flickr.com/assets/client-code.js" charset="utf-8"></script>

````python
def phi_flyer(shape, back=False):
    scalefactor = one/PHI # linear sf
    if not back:
        return shape * scalefactor    # shrink
    return shape * (one/scalefactor)  # grow

def gapper(shape, back=False):
    scalefactor = (2 / 3)**(one / 3) * (PHI / rt2(2)) # linear sf
    if not back:
        return shape * scalefactor   # shrink
    return shape * (one/scalefactor) # grow

def swell_bell(shape, back=False):
    scalefactor = (3 / 2)**(one / 3) # linear sf
    if not back:
        return shape * scalefactor   # grow
    return shape * (one/scalefactor) # shrink

def express(shape, back=False):
    scalefactor = 18**(-one/6)
    if not back:
        return shape * scalefactor   # shrink
    return shape * (one/scalefactor) # grow    
````

Below is some somewhat advanced Python wherein we define a Compasable class to enchance the capabilities of a function type object. By "swallowing" a particular function when birthing an instance of Composable, the function remains callable, even though it's now internal to the Composable object. Furthermore, Composable defines `__mul__` and `__rmul__` to return the result of composing the current object (with its internal function) with a composable on the left or on the right, such that `(f * g)(x)` is the same as `f(g(x))` and `(g * f)(x)` returns `g(f(x))`. Furthermore, `h = f * f * f` may instead be written `h = f ** 3`, to achieve the same result when for `h(x)`.

Then, once Composable is defined, we use it as a "decorator class" above the functions and inverse functions, using the `@` operator, which triggers precisely this swallowing action, with the resulting instance receiving the same name. From the "outsider" viewpoint, a decorated function seems almost industingishable from the original, in terms of how it's called, except now it has these added superpowers: of being able to multiply with (compose with) other functions; of being able to multiply with itself using `**` (the exponentiation operator).

Related Notebooks:

* [Composing Function Type Objects](https://github.com/4dsolutions/Python5/blob/master/ComposingFunctions.ipynb)
* [Decorating Objects](https://github.com/4dsolutions/Python5/blob/master/Abducted!.ipynb)
* [All Aboard the S Train](https://github.com/4dsolutions/Python5/blob/master/S_Train.ipynb)

````python
import types  # <-- to get FunctionType

class Composable:
    """
    Composable swallows a function, which may still be called, by
    calling the instance instead.  Used as a decorator, the Composable
    class enables composition of functions by means of multiplying
    and powering their corresponding Composable instances.
    """
    
    def __init__(self, func):
        self.func = func     # eat a callable
        
    def __call__(self, x):
        return self.func(x)  # still a callable
        
    def __mul__(self, other):
        """
        multiply two Composables i.e. (f * g)(x) == f(g(x))
        g might might a function. OK if f is Composable.
        """
        if isinstance(other, types.FunctionType): # OK if target is a function
            other = Composable(other)
        if not isinstance(other, Composable): # by this point, other must be one
            raise TypeError
            
        return Composable(lambda x: self.func(other.func(x)))  # compose 'em
    
    def __rmul__(self, other): # in case other is on the left
        """
        multiply two Composers i.e. (f * g)(x) == f(g(x))
        f might might a function. OK if g is Composer.
        """
        return Composable(other).__mul__(self)
        
    def __pow__(self, exp):
        """
        A function may compose with itself why not?
        """
        # type checking:  we want a non-negative integer
        if not isinstance(exp, int):
            raise TypeError
        if not exp > -1:
            raise ValueError
        me = self
        if exp == 0: # corner case
            return Composable(lambda x: x) # identify function
        elif exp == 1:
            return me                # (f**1) == f
        for _ in range(exp-1):       # e.g. once around loop if exp==2
            me = me * self
        return me
        
    def __repr__(self):
        return "Composable({})".format(self.func.__name__)
````

````python
@Composable
def phi_flyer(shape):
    scalefactor = one/PHI       # linear sf
    return shape * scalefactor  # shrink

@Composable
def inv_phi_flyer(shape):
    scalefactor = one/PHI             # linear sf
    return shape * (one/scalefactor)  # grow
    
@Composable
def gapper(shape):
    scalefactor = (2 / 3)**(one / 3) * (PHI / rt2(2)) # linear sf
    return shape * scalefactor   # shrink

@Composable
def inv_gapper(shape):
    scalefactor = (2 / 3)**(one / 3) * (PHI / rt2(2)) # linear sf
    return shape * (one/scalefactor) # grow
````

````python
day_trip = phi_flyer * inv_phi_flyer # round trip, shrink to RT_E then "grow home" to Union Station
````

````python
new_rt = day_trip(rt) # round trip
new_rt.volume
````

````python
express = phi_flyer * gapper  # compose express route from Union Station to RT_T
````

````python
new_rt = express(rt)
new_rt.volume
````

# Recalibrating the R-radius Sphere

```python
from IPython.display import YouTubeVideo
```

```python
from flextegrity import Tetrahedron, Icosahedron, Octahedron, Amod
from sympy import Integer, Rational, sqrt
import sympy
```

To help with recall, lets instantiate a specialcase octahedron using the defaults provided in the flextegrity module. 

```python
oc = Octahedron()
```

The `vertexes` dict reminds us of the whole non-negative coordinate addresses of IVM balls, six of which define the vertices of our Octahedron.

```python
oc.vertexes
```

These vertices surround (0,0,0,0), the origin, as Coupler tips, six of them, each the same distance away.

```python
oc.vertexes['i'].length()
```

Per our defaults, the A + B vector, the sum of quadrays A and B i.e. (1,0,0,0) + (0,1,0,0), pokes through the mid-edge of the homebase tetrahedron in the +Z direction.

```python
oc.vertexes['i'].xyz
```

Here's how we compute IVM ball lattice density, the ratio of the space occupied by the balls, to all of space. Since very IVM ball nests inside of, and is tangent to, a containing rhombic dodecahedron of volume 6, we have a way of capturing this ratio.

The native IVM volume of an IVM ball is obtained using the Synergetics Constant (S3) as a conversion constant. The rhombic dodecahedron has a native IVM volume (tetravolume) of six. So...

<a data-flickr-embed="true" href="https://www.flickr.com/photos/kirbyurner/55142463580/in/dateposted/" title="ivm_vs_xyz"><img src="https://live.staticflickr.com/65535/55142463580_a784e5d730.jpg" width="500" height="388" alt="ivm_vs_xyz"/></a><script async src="//embedr.flickr.com/assets/client-code.js" charset="utf-8"></script>

```python
S3 = sqrt(Rational(9,8))  # 1.06066...
density = sympy.pi * Rational(4,3) * S3 / Integer(6)
density
```

```python
density.evalf(100)
```

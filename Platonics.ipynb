Lets take a look at those same poly guys by running code in this Notebook and by building a volumes dictionary.

```python
from flextegrity import Cube, Octahedron, RT, Amod
from flextegrity import Tetrahedron, InvTetrahedron, RD, PD, Icosahedron
import sympy
```

```python
ϕ = sympy.Symbol('ϕ')
```

```python
# dual pair
orange_guy = Tetrahedron()
black_guy = InvTetrahedron()

# dual pair
green_guy = Cube()
red_guy = Octahedron()

# dual pair
brown_guy = PD()
cyan_guy = Icosahedron()
```

```python
expr = cyan_guy.volume
expr.subs(sympy.S(1)/2 + sympy.sqrt(5)/2, ϕ)
```

```python
inventory = {} # empty dict

for shape in (orange_guy, black_guy, 
              green_guy, red_guy, 
              brown_guy, cyan_guy):

    # simplify and evalf the volume is an algebraic expression, else pass integers through as is
    if type(shape.volume) == int:
        vol_alg = shape.volume
        vol_num = shape.volume
    else:
        vol_alg = shape.volume.subs(sympy.S(1)/2 + sympy.sqrt(5)/2, ϕ)
        vol_num = shape.volume.evalf(20)
        
    inventory[shape.name] = (vol_alg, vol_num)

inventory
```

A pretty pandas Dataframe will give us our Platonics table.

```python
import pandas as pd
```

```python
platonics_table = pd.DataFrame({"Shape"             : list(inventory.keys()),
                                "Volume (algebraic)": [t[0] for t in inventory.values()],
                                "Volume (numeric)"  : [t[1] for t in inventory.values()]})
platonics_table                 
```

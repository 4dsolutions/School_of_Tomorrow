# Disclosing Decorators

[Disclosure Day](https://www.imdb.com/title/tt15047880/), the Spielberg film, had just come out in this summer of 2026. The ET was all grown up now, [a grampa ET](https://controlroom.blogspot.com/2026/06/disclosure-day-movie-review.html), in a wheelchair. My how time flies.

Per [longstanding practice](https://github.com/4dsolutions/Python5/blob/master/Abducted!.ipynb), I deliberately entangle "Abduction" (as in "by aliens") with "Decoration" (as in the Python computer language).

"To decorate" is a process whereby an unsuspecting bystander, let's say, is "put through the wringer" (figuratively) meaning another callable eats our callable and returns a transformed callable (⋈), but with the name unchanged. Even the original docstring may be left intact using 'functools.wraps'. The bystander still acts and talks pretty much as before. But something has changed.

In my online classes I'd give the backstory for the science fiction TV show [The Invasion](https://www.imdb.com/title/tt0460651/?ref_=mv_close), set in the swampy Florida everglades, where pythons prevail (appropriately -- these were Python classes), wherein the abductees had no clue that been eaten and recast as mere facsimiles of their former selves.

Playing off the Spielberg movie, in this code the innocent bystander is made to "click click click" much like the Emily Blunt character, the weather lady.

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Wed Jun 24 09:16:01 2026

@author: kirbyurner
"""
from functools import wraps # for passing on the docstring

talk_balloon = "Hi, my name is {}.\nThoughts: {}\n"

def UFO(abductee):
    """
    just mindining my own business
    """
    @wraps(abductee)
    def one_of_us(name, thoughts):
        thoughts = "Click Click Click" # brainwash!
        return talk_balloon.format(name, thoughts)
    return one_of_us

def ordinary_guy(name, thoughts):
    """
    just minding my own business
    """
    return talk_balloon.format(name, thoughts) 

# Before
print(ordinary_guy("Hayden", "La Dee Da"))
help(ordinary_guy)

# Being abucted!
@UFO
def ordinary_guy(name, thoughts):
    """
    just minding my own business
    """
    return talk_balloon.format(name, thoughts) 

# After
print(ordinary_guy("Hayden", "La Dee Da"))
help(ordinary_guy)
```
